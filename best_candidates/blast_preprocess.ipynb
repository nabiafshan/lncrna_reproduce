```python
import re
import pandas as pd
```

```python
pattern_query=r"(?:Query= )(.+)"
pattern_len=r"(?:Length=)([0-9]+)"
# pattern_sequences=r"  ([\w\d]+)\s+(\d+.{0,1}\d*)\s+(.+)\s*?"
pattern_sequence=r"\s*> ([\w\d]+)\s"
pattern_sc_ex_met=r"\s*Score = (.+) bits.*Expect = (.*),\s+Method: (.*)"
pattern_id_pos_gap=r"\s*Identities = (.*),\s+Positives = (.*),\s+Gaps = (.*)\s+"
pattern_frame=r"\s*Frame = (.*)\s*"
```

```python
re.findall(pattern_sequence,
           """ Score = 88.6 bits (218),  Expect = 5e-19, Method: Compositional matrix adjust.
 Identities = 60/89 (67%), Positives = 64/89 (72%), Gaps = 0/89 (0%)
 Frame = +2""")
```

```python
res_dict={'query':[],'length':[],'sequence':[],'score':[],'expect':[],'method':[],'identities':[],'positives':[],'gaps':[],'frame':[]}

with open("misannot_results (9).out", "r") as file:

    query=length=sequence=score=expect=method=identities=positives=gaps=frame = None
    
    for i,line in enumerate(file.readlines()):
        
        res_query = re.findall(pattern_query,line)
        if len(res_query) > 0: 
            query=length=sequence=score=expect=method=identities=positives=gaps=frame = None
            query = res_query[0]
#             print(query)
            continue
            
        res_len = re.findall(pattern_len,line)
        if len(res_len) > 0: 
            length = res_len[0]
#             print(length)
            continue

        res_sequence = re.findall(pattern_sequence,line)
        if len(res_sequence) > 0: 
            sequence = res_sequence[0]
#             print(sequence)
            continue
            
        res_sc_ex_met = re.findall(pattern_sc_ex_met,line)
        if len(res_sc_ex_met) > 0: 
            score, expect, method = res_sc_ex_met[0]
#             print(score, expect, method)
            continue
            
        res_id_pos_gap = re.findall(pattern_id_pos_gap,line)
        if len(res_id_pos_gap) > 0: 
            identities, positives, gaps = res_id_pos_gap[0]
#             print(identities, positives, gaps)
            continue
            
        res_frame = re.findall(pattern_frame,line)
        if len(res_frame) > 0: 
            frame = res_frame[0]
#             print(frame)
            res_dict['query'].append(query)
            res_dict['length'].append(length)
            res_dict['sequence'].append(sequence)
            res_dict['score'].append(score)
            res_dict['expect'].append(expect)
            res_dict['method'].append(method)
            res_dict['identities'].append(identities)
            res_dict['positives'].append(positives)
            res_dict['gaps'].append(gaps)
            res_dict['frame'].append(frame)
```

```python
df_res = pd.DataFrame(res_dict)
df_res
```
