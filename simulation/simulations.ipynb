```python
import pathlib
import pandas as pd

from utils import is_mislabelled, get_sim_flip_percent, get_sim_mean, get_sim_sd
```

```python
filepaths = list(pathlib.Path('../data/simulations').glob('*.*'))
```

```python
results = dict()
for this_file in filepaths:
    df = pd.read_csv(this_file)
    df = df[df['flip'] == 1]
    df['is_mislabelled'] = df.apply(lambda row: is_mislabelled(row), axis=1)
    fname = this_file.stem.split('_')[1:5]
    fname = '_'.join(fname)
    results[fname] = df['is_mislabelled'].value_counts(normalize=True).to_dict()
```

```python
simulations = pd.DataFrame(results).T.reset_index()
simulations['mean'] = simulations.apply(lambda row: get_sim_mean(row), axis=1)
simulations['sd'] = simulations.apply(lambda row: get_sim_sd(row), axis=1)
simulations['flip_percent'] = simulations.apply(lambda row: get_sim_flip_percent(row), axis=1)
simulations
```

```python
simulations.drop(columns=['index'], axis=1).groupby(['flip_percent', 'mean', 'sd']).mean()
```
